```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data = pd.read_csv('C:/Users/signi/Downloads/Linear Regression/Restaurant_Profit_Data.csv')
```

```python
data.head()
```

```python
ohe=pd.get_dummies(data['City'])
ohe
data = pd.concat([data.reset_index(drop=True),ohe],axis=1)
data
data = data.drop('City',axis=1)
data
```

```python
features = ["Miscellaneous_Expenses","Advertising","Chicago","Mumbai"]
response = ["Profit"]
# X is independent variables
# Y is dependent variable
X=data[features].values
y=data[response].values
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model = LinearRegression()
```

```python
model = model.fit(X_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
import statsmodels.api as sm
X_with_constant = sm.add_constant(X_train)
model = sm.OLS(y_train, X_with_constant)
results = model.fit()
results.params
results.summary()
```

```python
y_pred = model.predict(X_train)
df = pd.DataFrame(data = y_train ,columns =['y_test'])
df['y_pred'] = y_pred
df
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE: %.2f' % mean_squared_error(y_train, y_pred))
print('R^2: %.2f' % r2_score(y_train, y_pred))
```

```python
y_predict = model.predict(X_test)
```

```python
df = pd.DataFrame(data = y_test ,columns =['y_test'])
df['y_predict'] = y_predict
df
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('MSE: %.2f' % mean_squared_error(y_test, y_predict))
print('R^2: %.2f' % r2_score(y_test, y_predict))
```
